D. Remove unnecessary parts of the data sets (e.g. duplicate entries, dates beyond the desired range, etc).


```python
#dropping data we are not interested in as we are only interested in precipitation, locations, and dates.
bad_cols = ["ELEVATION", "DAPR", "MDPR", "SNOW", "SNWD", "DASF", "MDSF"]
df_seattle = df_seattle.drop(columns=bad_cols)
df_ny = df_ny.drop(columns=bad_cols)
```

```python
df_ny = df_ny.drop_duplicates(subset="DATE")
df_seattle = df_seattle.drop_duplicates(subset="DATE")
```

I removed the variables that we are not interested in such as snow,elevation, etc and kept only PRCP or Precipitation. I was debating on getting rid of Longitude and Latitude but it will be good to keep if we want to map it on a interactive map or something. I also got rid of duplicate dates since we will be comparing one date per station in each city.

E. Identify missing values in the data sets. Impute or remove NaN values, as appropriate.


```python
df_seattle.dropna(subset= ['PRCP'])
```

```python
df_ny.dropna(subset= ['PRCP'])
```

Since the main piece of information we are trying to gather is rainfall, it only makes sense to remove the rows where NaN is present only for PRCP ( Precipitation) since we do not want rows with null data

F. Join the Seattle and New York data frames, keeping only the date and precipitation values.



```python
df_merge = df_ny.merge(df_seattle, on="DATE", how="left")
df_merge
```

```python
df_ny = df_ny.drop_duplicates(subset="DATE")
df_seattle = df_seattle.drop_duplicates(subset="DATE")
df = df_ny[['DATE', 'PRCP']].merge(df_seattle[['DATE', 'PRCP']], on='DATE', how='left')
df
```

Above I have my merge of the two datasets using the merge function and then below it I made another table showing them merged by date and it shows the precipitation value for seattle and newyork which are currently labeled PRCP Y and X

G. Ensure that the data frame is in a tidy, or long, format.

```python
df_tidy = pd.melt(df, id_vars = 'DATE', var_name = 'CITY', value_name = 'PRCP')

df_tidy
```

```python
df_tidy.loc[df_tidy['CITY']=='PRCP_x', 'CITY'] = 'NYC'
```

```python
df_tidy.loc[df_tidy['CITY']=='PRCP_y', 'CITY'] = 'SEA'
```

```python
df_tidy
```

Above I tidied up the merged data by creating a new column called City and having it display the two datasets in a combined column and renamed PRCP_y and PRCP_x to NYC and SEA.

H. Rename the columns, if necessary, to follow best practices of being lowercase, snake_case, and understandable.



```python
mapping_dict = {'DATE':'date', 'CITY':'city', 'PRCP':'precipitation'}
```

```python
df_tidyrainfall = df_tidy.rename(columns=mapping_dict)
df_tidyrainfall
```

I. Create any derived variables that will be useful in the analysis.



```python
df_tidyrainfall['month'] = pd.DatetimeIndex(df_tidyrainfall['date']).month
df_tidyrainfall["year"] = df_tidyrainfall["date"].dt.year
df_tidyrainfall
```

Above I added variables that are specific to the month and year of each data point to make it easier for me if I want to go back and sort/ display any data analysis that I do for a specific year or month.
